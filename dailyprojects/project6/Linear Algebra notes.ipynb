```python
##Vector #ùë•‚ààùëÖùëõ - an ordered list of n features

x = np.array([1,2])
##Length/Norm - Distance from origin

norm = np.linalg.norm(x)

##unit vector - pure direction
unit = x/norm

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example curve arrays (replace with your own data)
xx = np.linspace(0.1, 2.4, 2)
yy = x*x*1.4

# Chosen points
a, b = 0.04, 1.2

# Compute f(a), f(b) by interpolation (or direct lookup if a,b ‚àà xx)
fa = np.interp(a, xx, yy)
fb = np.interp(b, xx, yy)


```

```python
# Secant slope
m = (fb - fa) / (b - a)

# Secant line values over the same xx grid
y_sec = m * (xx - a) + fa

# Plot curve + secant
plt.figure(figsize=(6,4))
plt.plot(xx, yy, label='f(x)')                  # your original curve
plt.plot(xx, y_sec, '--', color='red',         # the secant line
         label=f'Secant through x={a},{b}\n slope={m:.2f}')
plt.scatter([a, b], [fa, fb], color='black')    # mark the two points
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Secant Line on f(x)')
plt.grid(True)
plt.show()
```
