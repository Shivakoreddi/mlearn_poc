```python
##Flag tiny/huge datasets; watch RAM if you‚Äôll bootstrap 100¬†trees
##Shape & memory (rows¬†√ó¬†cols, bytes)
##How many rows and columns?
##Are there any memory-hogging columns (e.g., long strings)?
##Confirm you‚Äôve loaded the data correctly

# Example synthetic dataset
n = 100
df = pd.read_csv(r'/Users/shiva/PycharmProjects/mlearn_poc/dailyprojects/project8/fraudTrain.csv')
df.info(memory_usage='deep')
df.head()
```

```python
##Actions and approach for Step1 - Data snapshot
##because the data is large around 1gb,
##Large enough to slow grid‚Äësearch loops and EDA plots

##reduce the dtypes for columns - 
df_float = df.select_dtypes('float64').columns
df[df_float] = df[df_float].astype('float32')
df_int = df.select_dtypes('int64').columns
df[df_int] = df[df_int ].astype('int32')

##Unnamed: 0 looks like a row index -- Adds no signal, wastes RAM.

df = df.drop(columns=['Unnamed: 0'])

##trans_date_trans_time is object - convert to datetime
df['trans_ts'] = pd.to_datetime(df['trans_date_trans_time'])
df = df.drop(columns=['trans_date_trans_time'])

df.info(memory_usage='deep')

```

```python
df['hour'] = df['trans_ts'].dt.hour
df['dow']  = df['trans_ts'].dt.dayofweek
```

```python
##Drop high‚Äëcardinality IDs & direct PII
df = df.drop(columns=['cc_num','trans_num','first','last','street','city','state'])

##age feature

df['dob'] = pd.to_datetime(df['dob'])
df['age'] = (df['trans_ts'] - df['dob']).dt.days // 365
df = df.drop(columns=['dob'])
```

## 8Ô∏è‚É£ Leakage Probes

```python
##Columns that perfectly predict the label (timestamps, IDs) inflate accuracy
##To find features that "cheat" by containing information about the target.
##Perfect correlation with the target
##Timestamp or ID-based leak
##Categorical values that map 1:1 with the label

# Numeric correlation
corr = df.corr(numeric_only=True)['is_fraud'].abs().sort_values(ascending=False)
print("Top correlations with target:")
print(corr)

# Perfect categorical predictors
for col in df.select_dtypes('object'):
    mapping = df.groupby(col)['is_fraud'].nunique()
    if (mapping == 1).all():
        print(f"{col} may be a leakage source.")
```

```python
##Check for skewness 
skewness = df[numeric_cols].skew().sort_values(ascending=False)
print(skewness)
```

```python
##Normalize Skewed Columns (Log / Box-Cox / Yeo-Johnson)
##Option A: Log Transform (use for positive-only features)

df['amt_log'] = np.log1p(df['amt'])  # log(amt + 1)
```

```python
##Option B: Box-Cox / Yeo-Johnson (scikit-learn)

from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')  # use 'box-cox' if all values > 0
df_transformed = pd.DataFrame(pt.fit_transform(df[numeric_cols]), columns=numeric_cols)
```

```python
##Recheck skew after transform
transformed_skewness = df_transformed.skew().sort_values(ascending=False)
print(transformed_skewness)
```

```python
##When Should You Normalize?

##Model Type. -	Normalize?
##Linear Regression, Logistic Regression	- ‚úÖ Yes
##SVM, k-NN	 - ‚úÖ Yes
##Tree-based models (RF, XGBoost, etc.)	- ‚ùå Not required
##Neural Networks	- ‚úÖ Yes (also standardize)


##Note - Why Tree-Based Models (like Random Forest, XGBoost) don‚Äôt need normalization or scaling?
##Because trees split on feature thresholds, not on distances, magnitudes, or gradients ‚Äî so the actual scale of the feature doesn‚Äôt matter.

## What ‚Äúnormalization‚Äù does:
## 1. Makes features comparable in scale (e.g., 0‚Äì1 or z-score)

## 2. Helps models like Logistic Regression, SVM, KNN, and Neural Networks perform better ‚Äî because:

## 3. They use distance, dot products, or gradient descent

## 4. A large-valued feature (e.g., income) could dominate others (e.g., age) if unscaled
```
